### 2. 数据分析用到哪些数学？

- **概率**

概率是一个抽象概念。它的分母，是要把所有假设样本空间Ω取尽的一种抽象，而现实中这一般是很难办到的。

---

>概率为零的事件不一定是**不可能事件**，比如：
>
>a. 在任意实数区间\[a, b\]内随机取一点，它正好是某一值（比如$ (a + b)\over 2$）的概率
>
>b. 随手抛一次均匀硬币，它是立着的概率
>
>而**不可能事件**的概率一定为零！

---

概率论、统计学就是研究概率并应用的数学分支。数据分析中大量应用概率论与统计学知识，学习者应当掌握包括概率、频率、期望、均值、方差、协方差、贝叶斯变换等基本概念，并能进行统计描述和假设检验。

- **代数**

线性代数知识在数据处理中广泛应用，特别是在数据挖掘、机器学习、神经网络中使用频率极高，有志于从事数据分析专业工作的学习者应夯实自己的线代基础，包括矩阵基础知识、克莱姆法则、矩阵作为一种映射、特征值、特征向量、PCA等。

- **分析**

分析学是数学的主要分支之一，以微积分方法为基本工具，以函数（映射、关系等更丰富的内涵）为主要研究对象，以极限为基本思想的众多数学经典分支及其现代拓展的统称。

分析学在数据分析中的应用主要在机器学习部分，如反向传播、回归、激活函数、LOSS函数等，需要学习者掌握导数、偏导、梯度下降、$Beta$函数、$Taylor$展开等。

- **图论**

由于概率图模型的需要，学习者应掌握一些基本的图论知识，譬如完全图的路径数的计算方法。

## 第二部分 统计学速览##

#### .1 基本概念

**统计学**是从一组数据中抽象出有用信息以支持决策的原理和方法。统计学的研究领域或方法分为两部分：统计描述 Descriptive Statistics和统计推断 Inferential Statistics。

A **population （总体）** is the complete collection of all elements to be studied. The collection is complete in the sense that it includes all subjects to be studied.

A **parameter（参数）** is a numerical measurement describing some characteristic of a population.

A **sample（样本）** is a sub-collection of elements drawn from a population.

A **Statistic（统计量）** is a numerical measurement describing some characteristic of a sample.

#### 1.1 集中趋势
		
- 均值

*算术平均数*

用于度量数据的集中趋势，是最常见的集中趋势的度量，如同平衡点，易受极端数值影响 

$$ \overline{X} = \frac{\sum_{i=1}^{n}{x_i}}{n} = \frac{x_1 + x_2 + x_3 + ... + x_n}{n}$$

*几何平均数*

常用于计算百分比、比率、指数、增长率等指标的平均数。

一般来说，几何平均数总是小于或等于算术平均数。

$$ \overline{X} = \sqrt[n]{x_1\cdot x_2\cdot x_3\cdot\ldots\cdot x_n}$$

为了避免零增长率或负增长率，几何平均数的计算一般改为如下公式。中国的GDP增长率计算用的是实际GDP的几何平均数。

$$ g = \sqrt[(n-1)]{\frac{x_2}{x_1} \cdot \frac{x_3}{x_2} \cdot \ldots \cdot \frac{x_n}{x_{n-1}}} $$

*算术平均数和几何平均数的区别*

例：假定海淀区小学2004年入学人数为1万，2005年入学人数增加50％，而2006年减少50％。问从2004-2006年间入学人数平均变化情况是多少？

2005年的入学人数为1.5万；2006年的入学人数为0.75万。

年度变化的算术平均数为0，而年度变化的几何平均数为负13.4%。显然，几何平均数更适合度量入学人数在不同时期的变化。

*加权平均数*

加权平均数：不同比重数据的平均数。公式为：

$$ \overline{𝑋 ̄_𝑊}=\frac{(𝑊_1 𝑋_1+𝑊_2 𝑋_2+⋯𝑊_𝑁 𝑋_𝑁)}{(𝑊_1+𝑊_2+⋯+𝑊_𝑁)} $$

例如：一个学生某门课程期中成绩为72分，期末成绩为86分。期中和期末成绩分别占总成绩的40%和60％，问总成绩是多少？"

$$ \overline{𝑋 ̄_𝑊}=\frac{(72×0.4+86×0.6)}{(0.4+0.6)}=80.4$$

再如：

已知2000年三次产业从业人员的平均受教育年限分别为6.9年、7.8年和12.1年，问全体从业人员的平均受教育年限是多少？

不能用简单算术平均！

要先查找各个产业的从业人员数量，或比重。例如，知道2000年三次产业比重分别为50％、22.5％和27.5％。

全体从业人员的平均受教育年限是：

$$ 0.5*6.9+0.225*7.8+0.275*12.1＝8.5 $$

*调和平均数*

调和平均数：倒数的算术平均数的倒数。主要用来计算速度的平均。

公式：
$$ \overline{X_H} = \frac{1}{\frac{1}{N}(\frac{1}{X_1} + \frac{1}{X_2} + \ldots + \frac{1}{X_n})}$$

例如：一个学生阅读2页书，读第一页时的速度折合为每小时20页，读第二页时的速度折合为每小时40页，问该生平均每小时阅读速度是多少？

- 中位数

用于度量数据的集中趋势Central Tendency，为排序序列的中间值，如果 n 是奇数, 即为排序序列居中位置的观测值，如果 n 是偶数, 则为排序序列两个居中位置的观测值的均值，中位数在序列中的位置，不受数据极端值的影响

- 众数

用于度量数据的集中趋势，为出现次数最多的数值，可从排序数组中观察得到，不受极值的影响，一组数据可能没有众数或有多个众数，对数值数据和类型数据均适用

- 值域中点

用于度量数据的集中趋势，是最小和最大观测值的平均值，对数据的极端值非常敏感，常用于金融分析和气象预报。

$$ Midpoint = \frac{X_{min} + X_{max}}{2} $$

- 四分位数

用于度量数据的非集中趋势，把排序数据等分为四个区间，$Q_i$是第i个四分位数的位置

![](quartile.jpg)

Q1，first quartile， 第一四分位数是处于（n+1)/4 位置上的观测值。 25%的观测值比第一四分位数小。

Q2，second quartile,  第二四分位数就是中位数。处于2(n+1)/4 =（n+1)/2的位置上。50%的观测值比中位数小。

Q3。 third quartile, 第三四分位数是处于 3(n+1)/4位置上的观测值。75%的观测值比第三四分位数小。

- 中轴数

用于度量数据的集中趋势Central Tendency，是第一和第三四分位数的平均值，不受数据极端值的影响，Not Affected by Extreme Values

$$ Medial\_Axis = \frac{Q_1 + Q_3}{2}$$

#### 1.2 变异程度

- 全距（极差）Range

度量数据的离散程度Measure of Dispersion，最大和最小观测值之差，不考虑数据如何分布取决于数据的极端值

$$ Range = X_{max} - X_{min} $$

- 限制全距 Restricted range

指95％分位数与5％分位数之差。其定义式如下： 

$$ Restricted\_Range = X_{p95} - X_{p5} $$

- 四分位间距

度量数据的离散程度Measure of Dispersion，为第三和第一四分位数之差，

$$ Inter\_Quartile\_Range = Q_3 - Q_1 $$

包括排序数据处于中间位置50%的观测值

- 联合全距比率 Federal range ratio

指限制全距除以5％分位数。其定义式如下：

$$ Federal\_range\_ratio = \frac{X_{p95} - X_{p5}}{X_{p5}}$$

- 极值比

极值比为最大值与最小值之比，不考虑数据是如何分布的，只取决于数据的极端值。其公式为： 

$$ R = \frac{X_{max}}{X_{min}}$$

- 限制极值比

应用极值比公式时需要注意最小值不能为0。有时为了消除奇异值，可以将极值比改为限制性极值比。它比较的一个接近于最大值（但一般不等于最大值）的数据与一个接近最小值（但一般不等于最小值）的数据之比。例如，95％分位数与5％分位数之比，90％分位数与10％分位数之比，第三四分位数与第一四分位数之比等。 

$$ RR = \frac{X_{p95}}{X_{p5}} $$

- 方差和标准差

度量数据的离散程度Measures of Dispersion，是最为常用的变异程度的度量，涉及数据的分布状况，表示全部观测值相对于均值的平均变异程度。样本方差的计算公式如下：

$$ S^2 = \frac{\sum_{i=1}^{n}{(X_i-\overline{X})^2}}{n-1} = \frac{(X_1-\overline{X})^2 + (X_2-\overline{X})^2 + \ldots + (X_n-\overline{X})^2}{n-1} $$

当分母不是n-1而是N的时候，就是总体方差N。

标准差是方差的平方根：

$$ S = \sqrt{\frac{\sum_{i=1}^{n}{(X_i-\overline{X})^2}}{n-1}} = \sqrt{\frac{(X_1-\overline{X})^2 + (X_2-\overline{X})^2 + \ldots + (X_n-\overline{X})^2}{n-1}}$$

- 变异系数

度量相对离散程度（Measure of Relative Dispersion）,表现了标准差相对于均值的比率,用来比较两组或多组计量单位不同的数据的变异程度

$$ CV = (\frac{S}{\overline{X}}) \cdot 100\% $$

- 基尼系数

基尼系数是20世纪初意大利经济学家基尼根据洛伦茨曲线设计的判断收入分配平等程度的指标。收入分配越是趋向平等，基尼系数也越小，反之，收入分配越是趋向不平等，那么基尼系数也越大。联合国有关组织规定：若低于0.2表示收入绝对平均；0.2-0.3表示比较平均；0.3-0.4表示相对合理；0.4-0.5表示收入差距较大；0.6以上表示收入差距悬殊。

设实际收入分配曲线和收入分配绝对平等曲线之间的面积为A，实际收入分配曲线右下方的面积为B。并以A除以（A+B）的商表示不平等程度。这个数值被称为基尼系数或称洛伦茨系数。如果A为零，基尼系数为零，表示收入分配完全平等；如果B为零则系数为1，收入分配绝对不平等。该系数可在零和1之间取任何值。收入分配越是趋向平等，洛伦茨曲线的弧度越小，基尼系数也越小，反之，收入分配越是趋向不平等，洛伦茨曲线的弧度越大，那么基尼系数也越大。

$$ Gini = 1 - \sum_{i=0}^{n-1}{(X_{i+1} - X_i)(Y_i + Y_{i+1})} = \sum_{i=1}^{n-1}{X_i Y_{i+1} - \sum_{i=1}^{n-1}{X_{i+1} Y_i}}$$

其中，X为累计人数比率；Y为累计收入比率； 

也可以把基尼系数的计算公式写为：

$$ Gini = 1 - \sum_{i=1}^{n}{\frac{1}{n}\cdot[2\sum_{j=1}^{i}{\frac{X_j}{\sum_{k=1}^{n}{X_k}} - \frac{X_i}{\sum_{k=1}^{n}{X_k}}}]}$$

Gini系数的计算步骤：

Step1：排序

Step2：计算比重

Step3：计算累积比重

Step4：计算2倍的梯形面积

Step5：计算基尼系数

*基尼系数的性质*

基尼系数反映的是相对差异，不是绝对差异；

如果所有人的收入都有相同比例的增加或者减少，则基尼系数的数值保持改变；

如果所有人的收入都有相同数量的增加，则基尼系数的数值将变小。

基尼系数变小，则相对差异一定变小，但是绝对差异未必变小。

## 第四部分 假设检验 ##

### 1. 基础知识

#### 1.1 假设的类型 Types of Hypotheses

假设是对总体参数的一种推断，也就是说，我们希望通过观察样本的特征，来判断总体是否具有同样的特征。

> 总体参数：均值、比率和方差

> 进行分析前必须先识别参数

零假设 The Null Hypothesis


1.  陈述需要检验的假设
		例如: H0: $\mu$= 50
2.  零假设用 H0 表示
3.  代表“通常”的情形 “Business as usual”
4.  总是包含等号“=” （比如=, $\geq$, $\leq$)
5.  检验以“假定零假设为真”开始


备择假设 The Alternative Hypothesis


1.	备择假设为零假设的对立情况并且表明了H0被拒绝后的可得出的结论
	例如: H1:$\mu \neq 50$
2.	备择假设用H1表示 
3. 表示对“通常”情形挑战
    Captures the challenge to ‘Business as Usual’
4. 从不包含等号

*那么如何建立假设呢？*

总体的平均看电视时间为12小时吗？

$$\mu = 12 \qquad \mu \neq 12 \qquad H_0:\mu = 12 \qquad H_1: \mu \neq 12$$ 

总体平均看电视时间并非12小时吗？

$$\mu \neq 12 \qquad \mu = 12 \qquad H_0:\mu \neq 12 \qquad H_1: \mu = 12$$ 

一顶帽子的花费小于等于$20吗? 

$$ H_0:\mu\leq20 \qquad H_1:\mu>20$$

书店里平均消费大于$25吗? 

$$ H_0:\mu\geq25 \qquad H_1:\mu<25$$

*现实问题的假设设立 Identifying Hypotheses in Problems*

1. 检验的是总体参数，不是样本统计量
2. 两种假设具有互斥性和完备性                              mutually exclusive & exhaustive
3. 有时先设定备择假设较为容易
	
    如要检验总体平均年龄是否小于50岁？
    
    原问题的统计学表述为 $H_0: \mu\geq50$
    
    对立情况的统计学表述 $H_1: \mu <50$

#### 1.2 显著性水平$\alpha$ Level of Significance

1.	显著性水平常用$\alpha$ 表示，决定零假设为真情况下样本统计量不大可能的取值的区间，决定了检验的临界值和拒绝域的大小。                                     
2.	实际应用中，$\alpha$ 的值往往由研究人员在检验开始时选定，通常取决于出现第一类错误的代价。在第一类错误可能导致严重后果的情况下,取＝.05或者更小的值。
3. $1 - \alpha$ 表示置信水平

![](hypothis.png)

![](hypothis_2.png)

![](rejection.png)

#### 1.3 检验决策错误 Errors in Making Decision

第一类错误 Type I Error

否真错误, 后果往往较为严重

出现第一类错误的概率为 $\alpha$

等于显著性水平 Level of Significance

第二类错误 Type II Error

存伪错误, 出现第二类错误的概率为 $\beta$

类似法庭裁决：第一类错误如同是陪审团裁定被告有罪，可事实上
被告是无辜的。第二类错误如同被告被裁定无罪然而事实上有罪。

![](twoerror.png)

**注意：**

$\alpha$和$\beta$存在反向关系,不能同时降低两类错误。

*影响$\beta$的因素*

1.	总体参数的真值

假设值与真值之差越大，$\beta$值越小

2. 总体的标准差 $\sigma$

随着$\sigma$的减小，$\beta$值减小

3. 显著性水平 $\alpha$

随着 $\alpha$值的增大，$\beta$值减小

4.	样本容量 n

随着样本容量的增大，$\beta$值减小

---

对于给定的$\alpha$值，增加样本容量会降低$\beta$值，从而提高检验能力

对于给定的样本，我们必须考虑两类错误之间此起彼伏的关系，降低出现第一类错误的风险会提高出现第二类错误的风险。

只有当$H_0$为真时才有置信度与第一类错误的概念；检验能力及第二类错误的风险只有当$H_1$为真时，才有意义。

---

### 2. 假设检验过程 Test Process

1）设定$H_0$ 和 $H_1$

2）选择检验方法和统计量

> 一般来说，可根据样本情况和检验目标来选择T检验、Z检验、F检验、$\chi^2$检验、ANOVA等五种，对应的统计量分别为T值、Z值、F值、$\chi^2$值、F值

3）选择显著水平 $\alpha$，确定拒绝域和接受域

4）计算样本统计量的值

5）根据判别准则作出判断

【例子】手动挡汽车与自动挡汽车的平均英里数（mileage）是否有显著的差别？

```python
cars_Automatic = cars[cars.GearType==' Automatic'].copy().reset_index()
cars_Manual = cars[cars.GearType==' Manual'].copy().reset_index()
```

```python
print("Average Mileage of Automatic cars:", cars_Automatic.Mileage.mean())
print("Average Mileage of Manual cars:", cars_Manual.Mileage.mean())
```

```python
#手动挡汽车英里数的平均值和标准差
print("Mean of Manual Cars:", cars_Manual.Mileage.mean())
print("Standard Deviation of Manual Cars:", cars_Manual.Mileage.std())
```

```python
#自动挡汽车英里数的平均值和标准差
print("Mean of Automatic Cars:", cars_Automatic.Mileage.mean())
print("Standard Deviation of Automatic Cars:", cars_Automatic.Mileage.std())
```

```python
#手动挡汽车英里数均值的置信区间
stats.norm.interval(0.95, loc=cars_Manual.Mileage.mean(), scale = cars_Manual.Mileage.std()/np.sqrt(len(cars_Manual)))
```

```python
#自动挡汽车英里数均值的置信区间
stats.norm.interval(0.95, loc=cars_Automatic.Mileage.mean(), scale = cars_Automatic.Mileage.std()/np.sqrt(len(cars_Automatic)))
```

```python
#效应量
print("Effect size:", cars_Manual.Mileage.mean() - cars_Automatic.Mileage.mean())
```

在这个例子中，我们可以进行t检验来进行均值比较。

t检验的零假设是：$\mu_1 = \mu_2$

也就是说，本例的零假设是：平均来看，手动挡车和自动挡车的英里数没有显著差别

下面我们进行t检验并确定p值

```python
stats.ttest_ind(cars_Manual.Mileage, cars_Automatic.Mileage, equal_var=True)
```

p值是效应量发生的概率，而在本例中，p值接近于0

结论：小概率事件竟然发生了，那么我们应该拒绝原假设，并认为手动挡汽车和自动挡汽车的平均英里数有显著差别。
